```python
import pandas as pd 
```

```python
df = pd.read_excel("Coffee Shop Sales.xlsx")
df.head(3)
```

```python
df.info()
```

# Total sales analysis

```python
df['mesiac'] = df['transaction_date'].dt.month

# nový stĺpec pre celkový výnos
df['total_sales'] = df['unit_price'] * df['transaction_qty']

# agregácia pre celkový výnos podľa mesiacov
mesiac_vynos = df.groupby('mesiac')['total_sales'].sum().reset_index()
mesiac_vynos

#% zmena 
mesiac_vynos['percent_change'] = ( (mesiac_vynos['total_sales'] - mesiac_vynos['total_sales'].shift(1)) /  mesiac_vynos['total_sales'].shift(1) * 100)

mesiac_vynos
```

```python
map_month = {
    1 : "Jan 2023",
    2 : "Feb 2023",
    3 : "March 2023",
    4 : "Apr 2023",
    5 : "Maj 2023",
    6 : "Jun 2023",
}

mesiac_vynos["mesiac"] = mesiac_vynos["mesiac"].map(map_month)
```

```python
mesiac_vynos
```

```python
mesiac_vynos.to_csv('nazov_suboru.csv', index=False)
```

# Total Order 

```python
total_order = df.groupby('mesiac')['transaction_id'].count().reset_index()
total_order
total_order["mesiac"] = total_order["mesiac"].map(map_month)
total_order
```

# Total Quantity Sold

```python
total_qty_sold = df.groupby('mesiac')['transaction_qty'].sum().reset_index()
total_qty_sold
total_qty_sold["mesiac"] = total_qty_sold["mesiac"].map(map_month)
total_qty_sold
```

```python
df["day"] = pd.to_datetime(df["transaction_date"]).dt.day_name()
```

```python
df["day_type"] = df["transaction_date"].dt.weekday.apply(lambda x: "Weekend" if x >= 5 else "Weekday")
df
```

# Sales by weekday/weekend

```python
df["month_for_order"] = df['mesiac'] = df['transaction_date'].dt.month
```

```python
map_month = {
    1: "januar",
    2: "februar",
    3: "marec",
    4: "april",
    5: "maj",
    6: "jun",
}

df.mesiac = df.mesiac.map(map_month)

sales_by_day_type = df.groupby(["day_type", "mesiac", "month_for_order"])["total_sales"].sum().reset_index()
sales_by_day_type['mesiac'] = pd.Categorical(sales_by_day_type['mesiac'], categories= map_month.values(), ordered=True)
sales_by_day_type = sales_by_day_type.sort_values(["day_type", "mesiac"])
sales_by_day_type
```

```python
sales_by_day_type.to_csv("tryit")
```
